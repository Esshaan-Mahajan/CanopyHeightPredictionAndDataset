NDVI

```python
img = ee.ImageCollection('LANDSAT/LE07/C01/T1_TOA').filterDate('2020-04-01', '2022-08-01')
def NDVI(image):
  return image.expression('float(b("B4") - b("B3")) / (b("B4") + b("B3"))')

ndvi = img.map(NDVI).mean()
print(ndvi.getInfo())
#u_lon = 4.8148
#u_lat = 45.7758
#u_poi = ee.Geometry.Point(u_lon, u_lat)
#scale =0.1
# ndvi = img.normalizedDifference(['sur_refl_b02', 'sur_refl_b01'])
# nd =ndvi.getRegion(u_poi,scale= 0.1).getInfo()
#nd = ndvi.sample(u_poi, scale).first().get('B4').getInfo()

#print('NDVI:', nd)
```

EVI

```python

def EVI(image):
  return image.expression('float(2.5*((b("B4") - b("B3")) / (b("B4") + 6*b("B3")-7.5*b("B1")+1)))')

evi = img.map(EVI).mean()
print(evi.getInfo())
#u_lon = 4.8148
#u_lat = 45.7758
#u_poi = ee.Geometry.Point(u_lon, u_lat)
#scale =0.1
# ndvi = img.normalizedDifference(['sur_refl_b02', 'sur_refl_b01'])
# nd =ndvi.getRegion(u_poi,scale= 0.1).getInfo()
#e = evi.sample(u_poi, scale).first().get('constant').getInfo()

#print('EVI:', e)

```
